```python
#Income Statement Program (COGS = 40% of Sales)
```

```python
sales=float(input("Enter Sales ($): "))
```

```python
cogs=0.40*sales
gross_profit=sales-cogs
```

```python
use_taxes=input("Do you want to calculate taxes & net income? (y/n): ").strip().lower()
```

```python
if use_taxes=="y":
  tax_rate_percent=float(input("Enter Tax Rate (%): "))
  tax_rate=tax_rate_percent/100

  taxes=gross_profit * tax_rate
  net_income= gross_profit -taxes
```

```python
print("\nINCOME STATEMENT")
print(f"Sales:           ${sales:,.2f}")
print(f"Cost of Goods Sold: ${cogs:,.2f}")
print(f"Gross Profit:    ${gross_profit:,.2f}")
print(f"Taxes:           ${taxes:,.2f}")
print(f"Net Income:      ${net_income:,.2f}")
```

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter an operator (+, -, *, /): ")

if operator == "+":
  result = num1 + num2
  print("Result:", result)
elif operator == "-":
  result = num1 - num2
  print("Result:", result)
elif operator == "*":
  result = num1 * num2
  print("Result:", result)
elif operator == "/":
  if num2 != 0:
    result = num1/num2
    print("Result:", result)
  else:
    print("Error: Cannot divide by zero")
else:
  print("Invalid operator")
```

```python
sales = float(input("Enter sales value ($): "))

for tax_rate in range(20, 41, 10):
  cogs=0.4*sales
  gross_profit=sales-cogs
  taxes=gross_profit*tax_rate/100
  net=gross_profit-taxes

  print(f"Sales: {sales}$")
  print(f"Tax rate: {tax_rate}%")
  print(f"Cost of Goods Sold: {cogs}$")
  print(f"Gross Profit: {gross_profit}$")
  print(f"Taxes: {taxes}$")
  print(f"Net Income: {net}$")
  print()
```

```python
import math
print("Ads\tProfit")

for ads in range(0,201,25):
  attendees=20+2*math.sqrt(ads)
  revenue=attendees*10
  profit=revenue-200-ads
  print(f"{ads:.1f}\t{profit:.1f}")
```

```python
import math

ticketPrice=10
defaultAudience=20

print("Ad$ Profit")
for adMoney in range(0,201,25):
  revenue= (defaultAudience + 2 * round(adMoney ** 0.5)) * ticketPrice
  costs=200+adMoney
  profit=revenue-costs
  print(f"{float(adMoney)}\t{float(profit)}")
```

```python
#replacing for loop with while loop
import math
print("Ads\tProfit")
ads=0 #define ads
while ads <=200:
  attendees=20+2*math.sqrt(ads)
  revenue=attendees*10
  profit=revenue-200-ads
  print(f"{ads:.1f}\t{profit:.1f}")
  ads += 25
```

```python
#Rock paper scissors game: human v. computer
import random
play = input("Do you want to play Rock-Paper-Scissos? (yes/no): ").lower()
while play=="yes":
  user_move=input("Choose rock, paper, or scissors: ").lower()
  computer_move=random.choice(["rock", "paper", "scissors"])
  print("Computer choose:", computer_move)
  #Determine winner
  if user_move==computer_move:
    print("It's a tie!")
  elif (user_move == "rock" and computer_move == "scissors") or \
          (user_move == "paper" and computer_move =="rock") or \
          (user_move =="scissors" and computer_move =="paper"):
          print("You win!")
  else:
        print("Computer wins!")
#Ask if they want to play again
  play=input("Do you want to play again? (yes/no): ").lower()
print("Thanks for playing!")
```

```python
numbers=[] #empty list of numbers
num=float(input("Enter a number (0 to stop): ")) #keep asking for input until the user enters 0
while num != 0:
  numbers.append(num) #append each number to the list
  num=float(input("Enter a number (0 to stop): "))
numbers.sort(reverse=True) #so its in descending oder
print("Numbers in descending order:", numbers)
```
