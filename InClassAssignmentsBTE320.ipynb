# **Inventory Managemet**

```python
import math

#User Input
D = float(input("Enter annual demand (units/year): "))
S = float(input("Enter ordering cost ($/order): "))
H = float(input("Enter holding cost ($/unit/year): "))

#EOQ Calculation
EOQ = math.sqrt((2 * D * S) / H)

#Additional Results
orders_per_year = D / EOQ
cycle_days = 365 / orders_per_year
annual_ordering_cost = orders_per_year * S
annual_holding_cost = (EOQ / 2) * H
total_annual_cost = annual_ordering_cost + annual_holding_cost

#Output
print("\n--- EOQ Results ---")
print(f"EOQ:                  {EOQ:.2f} units/order")
print(f"Orders per year:      {orders_per_year:.2f}")
print(f"Reorder every:        {cycle_days:.1f} days")
print(f"Annual ordering cost: ${annual_ordering_cost:.2f}")
print(f"Annual holding cost:  ${annual_holding_cost:.2f}")
print(f"Total annual cost:    ${total_annual_cost:.2f}")
```

# **Numerical Calculator (if-elif-else statements):**

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter an operator (+, -, *, /): ")

if operator == "+":
  result = num1 + num2
  print("Result:", result)
elif operator == "-":
  result = num1 - num2
  print("Result:", result)
elif operator == "*":
  result = num1 * num2
  print("Result:", result)
elif operator == "/":
  if num2 != 0:
    result = num1/num2
    print("Result:", result)
  else:
    print("Error: Cannot divide by zero")
else:
  print("Invalid operator")
```

# **Dynamic dictionary; N distinct key:value pairs (while loop/ if-elif-else statements)**

```python
N = int(input("How many entries? "))
counts = {}
i = 0
while i < N:
    name = input("Enter name: ")
    if name not in counts:
        counts[name] = 1
    else:
        counts[name] += 1
    i += 1
print("Final dictionary:", counts)
```

# **Quality Control Manager (exceptions)**

```python
def getValues(n):
    assert isinstance(n, int), "n must be an integer"
    assert n > 0, "n must be positive"
    values = []
    i = 0
    while i < n:
        try:
            val = float(input(f"Enter value {i+1}: "))
            if val < 0:
              print("Warning: negative value, skipping.")
              continue
            values.append(val)
            i += 1
        except ValueError:
          print("Warning: invalid input, please enter a number.")
    return values

def calcAverage(alist):
    assert isinstance (alist, list), "Input must be a list"
    if len(alist) < 2:
        raise Exception("List must have more than 2 values.")
    trimmed = alist[2:]
    return sum(trimmed) / len(trimmed)

try:
  n = int(input("How many values? "))
  values = getValues(n)
  print(f"Average (excluding first two): {calcAverage(values):.2f}")
except Exception as e:
  print(f"Error: {e}")
```

# **Greeting Class**

```python
# Greeting class
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last
  def display(self, formal=False):
    if formal:
      print(self.last + ", " + self.first)
    else:
      print(self.first)
  def __str__(self):
    return self.first + " " + self.last

g = Greeting("ParisElla", "Sharma")
g.display(True) #formal
g.display(False) #informal
print(g) #uses __str__
```

# **Student class**

```python
class Student:
    def __init__(self, name, number):
        self.name = name
        self.number = number
        self.courses = []

    def get_name(self):
        return self.name

    def get_courses(self):
        return self.courses

    def set_name(self, new_name):
        self.name = new_name

    def __len__(self):
        return len(self.courses)

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)
            print(f"{self.name} has been enrolled in {course}.")
        else:
            print(f"{self.name} is already enrolled in {course}.")


#Test
s = Student("Paris", 12345)

s.enroll("Math")
s.enroll("Biology")
s.enroll("Math")

print(s.get_name())
print(s.get_courses())
print(len(s))

s.set_name("Paris S.")
print(s.get_name())
```

# **Classes Point and Rectangle**

```python
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

class Rectangle:
  def __init__(self, p1, p2, p3, p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4

  def get_width(self):
    return self.p1.distance(self.p2)

  def get_height(self):
    return self.p2.distance(self.p3)

  def get_area(self):
    return self.get_width() * self.get_height()
```

```python
p1 = Point(0, 0)
p2 = Point(4, 0)
p3 = Point(4, 3)
p4 = Point(0, 3)

rect = Rectangle(p1, p2, p3, p4)
print(rect.get_width())
print(rect.get_height())
print(rect.get_area())
```

# **Class Employee - Manager**

```python
class Employee:
    def __init__(self, name, wage, hours):
        self.name = name
        self.__hourly_wage = wage
        self.hours = hours

    def get_wage(self):
        return self.__hourly_wage

    def set_wage(self, newWage):
        if newWage <= 0:
            print("Negative wage values not allowed.")
        else:
            self.__hourly_wage = newWage

    def calculate_pay(self):          # ← un-indented, now a real method
        return self.__hourly_wage * self.hours

    def __str__(self):                # ← replaced display_info with __str__
        return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}"


class Manager(Employee):
    def __init__(self, name, wage, hours, bonus):
        super().__init__(name, wage, hours)
        self.bonus = bonus

    def calculate_pay(self):
        return (1 + self.bonus) * self.get_wage() * self.hours


employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary is: {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is: {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```
